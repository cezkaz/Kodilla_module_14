```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', 
                    header=0)
df = data[0]
df.head()
```

```python
df.rename(columns={'TITLE':'TYTUŁ', 
                'ARTIST':'ARTYSTA', 
                'YEAR':'ROK', 
                'HIGH POSN':'MAX POZ'}, inplace=True)
df
```

```python
df['ARTYSTA'].unique()
```

```python
df['ARTYSTA'].value_counts()
```

    najczęściej na liście pojawiają się COLD PLAY i TAKE THAT po 3 razy

```python
df.sort_values(by='ARTYSTA')
```

```python
df['ARTYSTA'].count()
```

```python
df['ARTYSTA'].nunique()
```

    liczba pojedynczych artystów, którzy znajdują się na liście to 47

```python
df.rename(str.capitalize, axis='columns', inplace=True)
df
```

```python
df.drop('Max poz', axis=1)
```

```python
df.drop('Max poz', axis=1, inplace=True)
```

```python
df
```

```python
df.sort_values(by='Rok')
```

```python
df.groupby(['Artysta','Tytuł']).agg(Najwcześniej_wydany_album=('Rok','min'))
```

```python
df['Rok'].value_counts()
```

    Najwięcej albumów znajdujących się na liście wyszło w 2000 r. oraz 1987 r. po 4 w każdym z nich

```python
start_date = 1960
end_date = 1990
mask = (df['Rok'] >= start_date) & (df['Rok'] <= end_date)
df = df.loc[mask]
df['Rok'].count()
```

    pomiędzy rokiem 1960 a 1990 wydano 22 albumy

```python
df['Rok'].min()
```

```python
df['Rok'].idxmin()
```

```python
df[df['Rok'] == df['Rok'].min()]
```

    najwcześniej wydanym albumem był THE SOUND OF MUSIC zespołu ORIGINAL CAST RECORDING wydany w roku 1965

```python
df.groupby(['Artysta','Tytuł']).agg(Najwcześniej_wydany_album=('Rok','min'))
```

```python
df.groupby(['Rok','Artysta','Tytuł']).agg(Najwcześniej_wydany_album=('Rok','min'))
```

```python
df.groupby(['Artysta']).agg(Najwcześniej_wydany_album=('Rok','min'), Nazwa_albumu=('Tytuł','unique'))
```

```python
df.groupby(['Artysta']).agg(Rok_wydania_albumu=('Rok','min'), Nazwa_albumu=('Tytuł','first'))
```

```python
albums_grouped = df.groupby(['Artysta']).agg(Rok_wydania_albumu=('Rok','min'), Nazwa_albumu=('Tytuł','first'))
albums_grouped.to_csv('albums_grouped.csv', index=False)
```
