```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df = pd.DataFrame(meanBudgRevSortByYear,columns=['year','mean_budget','mean_revenue'])

fig = plt.figure()
ax1 = plt.subplot(1,1,1)
ax1.bar(df.year, df.mean_revenue, label='revenues')
ax2 = ax1.twinx()
ax1.plot(df.year, df.mean_budget,color='red', label='budget',linewidth=3)


ax1.legend(bbox_to_anchor=(1.05,1), loc='upper left', borderaxespad=0.)

ax2.legend(bbox_to_anchor=(1.05,1), loc='upper left', borderaxespad=0.)

plt.title('Średni przychód i budżet w latach 2010-2016')
plt.show()
```

```python
df = pd.DataFrame(meanBudgRevSortByYear,columns=['year','mean_budget','mean_revenue'])

fig = plt.figure()
ax1 = plt.subplot(1,1,1)
ax1.bar(df.year, df.mean_revenue, label='revenues')
ax2 = ax1.twinx()
ax1.plot(df.year, df.mean_budget,color='red', label='budget',linewidth=3)

red_patch = mpatches.Patch(color='red',label='revenues')
ax1.legend(handles=[red_patch])

blue_patch = mpatches.Patch(color='blue',label='budget')
ax2.legend(handles=[blue_patch])

plt.title('Średni przychód i budżet w latach 2010-2016')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from matplotlib.ticker import FuncFormatter
import matplotlib.ticker as ticker
%matplotlib inline

def million(x, pos):
    return ' {:2.1f}M'.format(x*1e-6)

df = pd.DataFrame(meanBudgRevSortByYear,columns=['year','mean_budget','mean_revenue'])

formatter = plt.FuncFormatter(million)
fig = plt.figure()
ax1 = plt.subplot(1,1,1)
ax1.bar(df.year, df.mean_revenue, label='revenues')
ax2 = ax1.twinx()
ax1.plot(df.year, df.mean_budget,color='red', label='budget',linewidth=3)

lines,labels = ax2.get_legend_handles_labels()
lines2,labels2 = ax1.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=(1.05,0.8))
ax1.yaxis.set_major_formatter(formatter)
plt.title('Średni przychód i budżet w latach 2010-2016')

plt.show()
```

## Ad.6
### Stwórz histogram czasu trwania filmów z gatunku, który cechuje się największym średnim czasem trwania.

```python
import datetime as dt
```

```python
gen = df_movies_genre_id['genres'].unique()
```

```python
gen
```

```python
myOwnGroup = {}
```

```python
for genre in gen:
    myOwnSubDataFrame = df_movies_genre_id.where(df_movies_genre_id['genres'] == genre).dropna()
    myOwnGroup[genre] = myOwnSubDataFrame
```

```python
myOwnGroupHistory = myOwnGroup['History']
```

```python
myOwnGroupHistory.head(10)
```

```python
myOwnGroupHistory['runtime'].plot(kind='hist',bins=15, title='Histogram czasu trwania filmów z gatunku History',width=15)
```
